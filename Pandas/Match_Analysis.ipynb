# isin()



```python
#find the most destructive death over batsman in the history
mask1=data1[data1['over']>15]
mask1
```

```python
all_batsman=mask1.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list= all_batsman[x].index.tolist() # tolist() convert index in list
```

```python
#runs scored by all these 43 batsman and balls played by these 43 batsman

final= mask1[data1['batsman'].isin(batsman_list)]
final
```

```python
run=final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
sr=(run/balls)*100
sr
```

# merge() function same as sql join

```python
#syntex of merging two data frame
#       first table name.merge(the name of the 2nd data frame,left_on='column name',right_on =2nd column name)
new=data1.merge(data,left_on='match_id',right_on='id')
new
```

```python
print(new.shape)
print(data1.shape)
print(data.shape)
```

```python
new1=new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False)
 
# reset _index() is used for converting series into dataframe
new2=new1.reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
new2
```

```python
new2[['season','batsman']]
```
